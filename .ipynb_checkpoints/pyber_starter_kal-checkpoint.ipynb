```python
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import pandas as pd
import numpy as np

# Files to Load 
Pyber_city_data = '../Pyber_homework/data/city_data.csv'
Pyber_ride_data = '../Pyber_homework/data/ride_data.csv'


#read files
city_data_df = pd.read_csv(Pyber_city_data, low_memory=False)
ride_data_df = pd.read_csv(Pyber_ride_data, low_memory=False)

#Combine the data into a single dataset
combine_city_ride_df = pd.merge( ride_data_df,city_data_df,
                                 how='left', on='city')
combine_city_ride_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
#### Obtain the x and y coordinates for each of the three city types

#group the df by city
city_ride_group = combine_city_ride_df.groupby('city' )
 
#calculate average fare , ride count 
combine_mean = round(city_ride_group["fare"].mean(),0)
combine_count = city_ride_group["ride_id"].count()
city_type = city_ride_group["type"].max()
driver_mean = city_ride_group["driver_count"].max()

#create a summary df
city_ride_df = pd.DataFrame({"Average Fare":combine_mean ,"Ride Count":combine_count,
                             "city type":city_type,"driver_mean":driver_mean})

#split df by city type
Urban_type = city_ride_df.loc[city_ride_df['city type'] == 'Urban']
Suburban_type = city_ride_df.loc[city_ride_df['city type'] == 'Suburban']
Rural_type = city_ride_df.loc[city_ride_df['city type'] == 'Rural']

#Build the scatter plots for each city types

Urban = plt.scatter( x=Urban_type["Ride Count"], y=Urban_type["Average Fare"], marker="o",s=Urban_type["driver_mean"]*10,facecolors="lightcoral", edgecolors="black", alpha=0.70)
Suburban = plt.scatter( x=Suburban_type["Ride Count"], y=Suburban_type["Average Fare"], s=Suburban_type["driver_mean"]*10,marker="o",facecolors="lightskyblue", edgecolors="black", alpha=0.70)
Rural = plt.scatter( x=Rural_type["Ride Count"], y=Rural_type["Average Fare"], s=Rural_type["driver_mean"]*10,marker="o",facecolors="gold", edgecolors="black", alpha=0.70)


# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data(2016)")
plt.xlabel("Total Number of Rides(Per City)")
plt.ylabel("Average Fare($)")
plt.grid()

# Create a legend
lgnd = plt.legend([Urban,Suburban,Rural],['Urban', 'Suburban', 'Rural'], title="City Types" ,loc="best")
for handle in lgnd.legendHandles:
    handle.set_sizes([30.0])

# Incorporate a text label regarding circle size
plt.text(42,35, 'Note:\nCircle Size correlates with driver count per city' )

# Save Figure
plt.savefig("../Pyber_homework/Images/PyberRideSharing.png")

```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

city_type_group = combine_city_ride_df.groupby("type")

fare_type_count = city_type_group["fare"].sum()


# Build Pie Chart

# Labels for the sections of our pie chart
labels = ["Rural","Suburban","Urban"]

# The colors of each section of the pie chart
colors = [ "gold","lightskyblue","lightcoral",]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0, 0, 0.1)

plt.pie(fare_type_count, explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=150)

plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("../Pyber_homework/Images/Pyber_fare_type_pie.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Type Percents

ride_type_count = city_type_group["ride_id"].count()

# Build Pie Chart

# Labels for the sections of our pie chart
labels = ["Rural","Suburban","Urban"]

# The colors of each section of the pie chart
colors = [ "gold","lightskyblue","lightcoral",]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0, 0, 0.1)

plt.pie(ride_type_count, explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=150)

plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("../Pyber_homework/Images/Pyber_ride_type_pie.png")
```

```python
# Show Figure
plt.show()
```
